De todos modos, la gente tiende a encontrar formas de eludir las restricciones impuestas por los altos precios. Una forma muy popular de tratar de gastar menos dinero en un restaurante sin empeorar la experiencia del restaurante de manera significativa, es ir con un grupo de amigos y compartir un entrante. Esto suena como una técnica efectiva, especialmente si ves los restaurantes como un tipo de experiencia de "pasar el rato". Por supuesto, el punto no es compartir un aperitivo independientemente del tamaño del grupo. Esto negaría la idea de no empeorar la experiencia, en su lugar intentaremos reducir la cantidad de platos por persona solo hasta cierto punto. Creo que este punto es la mitad de la cantidad de personas en el grupo. De esta manera, nunca pagarás menos de la mitad del costo del plato, pero siempre comerás al menos la mitad. Asi, un grupo de dos compartirá un entrante, un grupo de tres o cuatro compartirá dos, un grupo de cinco o seis compartirá tres, y así sucesivamente. Debido a que es una práctica tan común, considero interesante poner en perspectiva el recorte de gastos que esta permite.

Es por eso que pensé que podría usar el índice de diversidad de Simpson para clasificar los restaurantes en La Habana de acuerdo con cuán diverso es el servicio que ofrecen. El índice de diversidad de Los Simpson es una medida de probabilidad, indica la probabilidad de elegir dos elementos del mismo tipo dentro de un conjunto de datos cuando estos dos elementos se eligen al azar. 6. Pasta
7. Pizza
8. Sándwich/Hamburguesa/Perrito caliente
9. Ninguna de las anteriores

Hay algunos matices en este proceso.
